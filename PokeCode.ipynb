```python
# data_cleaning.py

import pandas as pd
from sklearn.cluster import KMeans

df = pd.read_excel('pokemon.xlsx')

df.columns = df.columns.str.strip()

df['Name'] = df['Name'].str.strip()
df['Type'] = df['Type'].str.strip()

df = df.drop_duplicates(subset=['Name', 'Total', 'HP', 'Attack', 'Defense', 'Special Attack', 'Special Defense', 'Speed'])

df['Total Stats'] = df[['Attack', 'Defense', 'Speed', 'HP', 'Special Attack', 'Special Defense']].sum(axis=1)

def speed_category(speed):
    if speed >= 100:
        return 'Fast'
    elif speed >= 50:
        return 'Average'
    else:
        return 'Slow'

df['Speed Category'] = df['Speed'].apply(speed_category)

df['Primary Type'] = df['Type']

df_sorted = df.sort_values(by='Total Stats', ascending=False)

df_sorted.to_csv('pokemon_cleaned.csv', index=False)

print("✅ Saved to'pokemon_cleaned.csv'!")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('pokemon_cleaned.csv')

plt.figure(figsize=(10,6))
plt.hist(df['Total Stats'], bins=30, edgecolor='black')
plt.title('Distribution Total Stats')
plt.xlabel('Total Stats')
plt.ylabel('Quantity of Pokemons')
plt.grid()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Attack'], df['Defense'], c=df['Speed'], cmap='coolwarm', alpha=0.7)
plt.colorbar(label='Speed')
plt.title('Attack vs Defence (Color = speed))')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.grid()
plt.show()
```

```python
top10 = df.sort_values(by='Total Stats', ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.barh(top10['Name'], top10['Total Stats'], color='purple')
plt.xlabel('Total Stats')
plt.title('Top-10 by Power')
plt.gca().invert_yaxis()
plt.grid()
plt.show()
```
