```python
# data_cleaning.py

import pandas as pd
from sklearn.cluster import KMeans

# Загружаем данные
df = pd.read_excel('pokemon.xlsx')

# Очистка названий колонок от лишних пробелов
df.columns = df.columns.str.strip()

# Очистка названий покемонов и типов
df['Name'] = df['Name'].str.strip()
df['Type'] = df['Type'].str.strip()

# Удаление дубликатов (если покемон с несколькими типами - оставляем только одну запись)
df = df.drop_duplicates(subset=['Name', 'Total', 'HP', 'Attack', 'Defense', 'Special Attack', 'Special Defense', 'Speed'])

# Создание нового признака "Total Stats" (сумма основных боевых характеристик)
df['Total Stats'] = df[['Attack', 'Defense', 'Speed', 'HP', 'Special Attack', 'Special Defense']].sum(axis=1)

# Создание категорий скорости (опционально для дальнейшего анализа)
def speed_category(speed):
    if speed >= 100:
        return 'Fast'
    elif speed >= 50:
        return 'Average'
    else:
        return 'Slow'

df['Speed Category'] = df['Speed'].apply(speed_category)

# Обработка типа (если нужно сделать один главный тип)
# Например, оставить только основной тип (Type1)
df['Primary Type'] = df['Type']

# Сортировка по Total Stats для проверки
df_sorted = df.sort_values(by='Total Stats', ascending=False)

# Сохраняем обработанные данные в новый файл
df_sorted.to_csv('pokemon_cleaned.csv', index=False)

print("✅ Данные успешно обработаны и сохранены в 'pokemon_cleaned.csv'!")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка очищенных данных
df = pd.read_csv('pokemon_cleaned.csv')

# 1. Распределение Total Stats
plt.figure(figsize=(10,6))
plt.hist(df['Total Stats'], bins=30, edgecolor='black')
plt.title('Распределение Total Stats у покемонов')
plt.xlabel('Total Stats')
plt.ylabel('Количество покемонов')
plt.grid()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Attack'], df['Defense'], c=df['Speed'], cmap='coolwarm', alpha=0.7)
plt.colorbar(label='Speed')
plt.title('Атака vs Защита у покемонов (цвет = Скорость)')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.grid()
plt.show()
```

```python
top10 = df.sort_values(by='Total Stats', ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.barh(top10['Name'], top10['Total Stats'], color='purple')
plt.xlabel('Total Stats')
plt.title('Топ-10 покемонов по общей силе')
plt.gca().invert_yaxis()
plt.grid()
plt.show()
```
